# **How We Tell Stories:  Exploring AI and Human Interpretation of a 19th Century Continental Tour through Textual and Visual Reconstructions**

## Introduction

This project investigates how storytelling of cultural heritage objects changes when humans and AI are involved in interpreting and re-producing content.

Based on sketches and images from [Andrew Macgeorge's Journal of a Continental Tour 1867-1868](https://www.gla.ac.uk/collections/#/details?irn=250262&catType=C&referrer=/results&q=ms+gen+13), the project is an interactive and experimental process that shifts between image and language, designed as an exhibit for public engagement with the book.

The participant is invited, alongside AI, to describe, redraw, and interpret illustrations from the Journal. Through this journey, they are invited to question how the meaning of the stories Macgeorge tells through his drawings changes as it gets retold in different mediums and by different storytellers.

### Objectives of this project

Through the creation of an interactive exhibit, this project aims to engage a general audience with the book (a cultural heritage object) in a way that prompts users to reflect on both human and AI creative processes.

By tasking the user and the AI to simultaneously caption images and draw based on the same description, the experience engages with Macgeorge’s drawings to recreate and emulate them.

Using generative AI models such as Stable Diffusion for image generation and vision‑language models such as BLIP (Bootstrapped Language-Image Pretraining) for captioning, the outputs can be compared with the users’ work.

There are also AI-human collaborations where AI will generate images based on the users’ captions and describe the users’ drawings. All with the purpose of raising critical questions about how humans and AI understand the meaning of Macgeorge’s work as it goes through these processes and how this can change or alter his stories.

This is not to test machine accuracy or human drawing ability, but to see how the interaction can become an artistic form itself. Does it enhance the original, does it erase the narrative, or does it produce something completely new of value?

It will be broken down into conversational steps, with two stages. The first is seeing, saying, and seeing, and the second is reading, drawing, and seeing. The results of each will offer opportunities to explore discussions on the skills, knowledge, and background that influence the creator, and on how co-creation alters the creative process.

## Methodology

The two structured stages provide complimenatary translations of visual and textual story telling, putting the user in the centre of creation. A demontrative notebook [2_HowWeTellStories_AI_Human_Demonstration.ipynb](2_HowWeTellStories_AI_Human_Demonstration.ipynb) has been created to illustrate how the exhibition would work and the kinds of models that could be used.

Below is an outline of these processes that expands on the purpose of each step and how it meets my objectives.

### STAGE 1: AI and human captioning and reconstruction of the same image

The AI and the user both look at the same drawing from Journal of a Continental Tour (1867-1868) and generate their own textual descriptions.  
Each description is then used as a prompt for image generation, producing two reconstructed versions that can be compared with the original drawing.  

>_This stage explores how both human and machine “see” the same image and how transition between images and language reshapes meaning._

---

#### **Overview**

1. **Input selection**  
   A drawing from the book is chosen as the shared reference.  
   - Exhibit: Visitors select from a curated gallery of scanned illustrations.  
   - Code: In Colab, this is simulated with an image upload.

2. **AI description**  
   The BLIP image-captioning model analyses the drawing and produces a description generated purely from visual features.  
   - This uses the discriminative stage of the ML workflow, turning image into text prediction.

3. **Human description**  
   The participant writes their own caption for the same image.  
   - Provides their own subjective, cultural, or emotional interpretation.
   (unique to humans, something AI cannot provide)

4. **Reconstruction through generation**  
   Each caption (AI + human) becomes a prompt for the Stable Diffusion model, which creates new images from text input.  
   - Demonstrates the generative stage of the ML workflow where text is becomes an image.

5. **Comparison and reflection**  
   The three outputs, the original drawing, the AI-generated reconstruction, and the human-generated reconstruction are displayed together.  
   Participant is invited to consider:
   - How similar are they to the original?
   - Where have new narratives appeared?  
   - How does the process of translation alter the story?

---

#### **Key takeaways**

- Movement between media (image, text and image) as a metaphor for storytelling and interpretation.  
- Differences between machine perception (pattern recognition) and human perception (context and background).  
- Reflection on the value and creativity of both human and AI storytellers.


### STAGE 2: AI and human image generation from the same description

This stage reverses the process of Stage 1.  
Both the AI and the user receive the same caption derived from an (unseen) drawing in the journal.  
Each creates an image based solely on that written description.  
When the authentic original image is finally revealed, the participant can compare all three versions and reflect on how verbal prompts shape imagination.

---

#### **Overview**

1. **Input description**  
   A pre-written textual description based on a sketch from the book.
   - Exhibit: The text is displayed to the user, but the source image remains hidden.  
   - Code: Stored as a string variable in the notebook.

2. **AI image generation**  
   The Stable Diffusion model turns the description into a new image.  


3. **Human drawing**  
   The user creates their own image in response to the same description.  
   - Exhibit: Visitors sketch digitally or on paper, then photograph or upload their drawing.  
   - Code: A simple file upload simulates this step.

4. **Original image revealed**  
   The real 19th-century drawing is displayed at the end.  
   - To enable direct visual comparison across each interpretation.

5. **AI describes the human drawing**  
   The BLIP captioning model generates a textual description of the human artwork.  
   - This closes the interpretive loop, while also adding new layer of creative complexity.
   - Model may not identify drawing well, depending on users drawing style. Ask user to reflect what it means to have their work defined by an AI.

---

#### **Key takeaways**

- Reverses the flow of Stage 1 to observe the imaginative gap between language and imagery.  
- Highlights how textual cues can constrain or inspire different visual outputs.  
- Reveals how both human and machine reconstruct meaning from minimal verbal information.  
- Encourages reflection on creative authorship, relationship to source materials, and the evolution of a story through successive mediations.


### How the models work

#### Image Generation - Stable Diffusion

To generate images in this project, I used Stable Diffusion from the [Hugging Face Diffusers Library](https://huggingface.co/docs/diffusers/api/pipelines/stable_diffusion/text2img).

This is a [text-to-image](https://huggingface.co/docs/diffusers/api/pipelines/stable_diffusion/text2img) generation model that was origninall developed by  [CompVis](https://github.com/CompVis), [Stability AI](https://stability.ai/), [Runway](https://github.com/runwayml), and [LAION](https://laion.ai/).

It has been trained on 512x512 images from the LAION-5B dataset, which is a large collection of web-scraped images with associated captions. It is open source Latent Diffusion Model (LDM), which means it uses latent space instead of pixel space and can be run through Google Colab using standard GPUs.

Stable Diffusion works by learning to invert noise to create images. During training, it constantly adds and removes random noise from images, enabling it to recreate them accurately. A text encoder from CLIP connects words with visual features, allowing the model to link language to imagery. When generating images, it starts with pure noise in a compressed “latent space” and gradually refines it based on the text prompt, resulting in a clear image. This latent image is then decoded back into a standard 512 × 512 pixel picture using an autoencoder, producing the final image for viewers.

---

Learn more about how Diffusers work in [Hugging Face's libary for diffusion models](https://colab.research.google.com/github/huggingface/notebooks/blob/main/diffusers/diffusers_intro.ipynb#scrollTo=wW8o1Wp0zRkq).

> This model was chosen for this project because it is open source and can work directly between Colab and Hugging Face without the cost of APIs. It only requires a basic GPU, which Colab's T4 provides, and produces results within a few minutes. It was used across stages 1 and 2 to ensure that any differences in results were due to differences in captions, not to the AI's capabilities.

##### Limitations

As Stable Diffusion relies on a compressed latent space and was trained on uncurated web data, which is not specialised in 1900s sketch art, its results often show soft focus and distorted anatomy. However, these results are meaningful to the project as they reveal the statistical, rather than perceptual, nature of the AI’s ‘interpretation.’ The contrast between human and machine depictions of the same captions highlights how representation and storytelling differ when driven by probabilistic learning compared to the user’s lived experience.

#### Image Captioning - BLIP

To generate image captions in this project, I used BLIP ([Bootstrapped Language-Image Pre-training](https://huggingface.co/Salesforce/blip-image-captioning-large)) from the [Hugging Face Transformers Library](https://huggingface.co/docs/transformers/v5.8.0/en/model_doc/blip#blip).
BLIP is a vision-language model originally developed by researchers at [Salesforce AI Research](https://www.salesforceairesearch.com/). It was trained on millions of image-text pairs collected from the web, learning how visual elements correspond to words and phrases. The model analyses the visual content of an image with a language decoder, which turns these features into natural language.

When an image is provided, BLIP's vision encoder extracts key patterns such as shapes, objects, and composition. These visual features are then passed to the language model, which predicts a descriptive sentence that best matches the scene. Through this image to text process, BLIP effectively “looks” at the image and produces a caption that reflects what it recognises.

Its open-source design and integration through Hugging Face make it accessible for experimentation and easy to run in Google Colab without needing to pay for access or specialised hardware.

---

> This model was ideal for this project as it as it directly connects visual understanding to language, reflecting the projects focus on how humans and AI interpret stories through Macgeorge's sketches.  BLIP is able to 'look' at an image and generate a caption to reflect what it sees, which can be easily compared to how a human would do the same task. As it was open source, it could easily be used in Google Colab without specialist software or paying for access.

##### Limitations

This model has been trained on a large web-scraped data set, similar to Stable Diffusion, so it is not specialised in 1900s artwork and cultural contexts. This also means it can inherit inconsistencies from its training data that might misidentify historical images. It produces statistical predictions, not reasoned interpretations. This model produces short, literal descriptions, which can lose the artistic nuance in the images. However, this ties into the project's objectives, as it offers insight into what a machine uses to interpret images versus humans.

## Final Reflections

The demonstration notebook showcases code that uses AI models to simulate a user's interaction with the exhibition. It offers an interactive experience that encourages collaboration and comparison with AI. Each execution generates different versions of Macgeorge's original works, illustrating how translating images into text can significantly alter their meaning. The focus isn't on ranking which is better—the AI or the human—but on understanding the factors that influence each in creating something novel. This project highlights the creative and cultural processes embedded within artistic judgment and algorithmic computation.  

The differences are often where the value lies. AI can produce far more complex, detailed, and colourful images in a short time by using models that classify, predict, and reconstruct, but humans can connect on an emotional level and potentially add their own perspective to the drawings or captions, thereby evolving the original meaning. Asking what the value of each iteration encourages the user to consider how it impacts storytelling, for what purposes a story is to be retold, and how the re-teller can alter the original meaning. This could add value by offering new perspectives on 19th-century travel, but it should also be assessed whether it indiscriminately removes the author's intended meaning.

In **Stage 1**, BLIP's captions captured objects and basic composition but overlooked artistic style and context, reflecting the model's training on web-scraped imagery. When Stable Diffusion redrew the captions, it produced polished but generic, sometimes distorted scenes. This demonstrated how AI fills gaps in understanding with patterns learned from its training data rather than with creative intent. By putting the human descriptions, despite their human interpretation, through the same AI generation, it highlights to the user how the AI may misunderstand their description and produce something unexpected.

**Stage 2** reversed the process, showing how language guides visual creation. From the same written description, AI- and human-produced images can diverge widely, with one governed by datasets and the other by imagination. This encourages the user to consider what informs their imagination, what they have been 'trained' on, what skills they have, and how these affect the drawing they create. Revealing the original drawing exposed how much interpretation occurs between words and images.
